```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Traffic": [1200, 1450, 1100, 1300, 1700, 2200, 1900],
    "Revenue": [12000, 15500, 10500, 14000, 19500, 26000, 21000],
    "Platform_Tier": ["Basic", "Basic", "Standard", "Standard", "Premium", "Premium", "Premium"]
}

df = pd.DataFrame(data)

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(df["Day"], df["Revenue"], color="blue", marker="o")

ax.set_title("Revenue Over 7 Days")
ax.set_xlabel("Day")
ax.set_ylabel("Revenue")
ax.grid(True)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Traffic": [1200, 1450, 1100, 1300, 1700, 2200, 1900],
    "Revenue": [12000, 15500, 10500, 14000, 19500, 26000, 21000],
    "Platform_Tier": ["Basic", "Basic", "Standard", "Standard", "Premium", "Premium", "Premium"]
}

df = pd.DataFrame(data)

plt.figure(figsize=(7,5))

sns.barplot(
    data=df,
    x="Platform_Tier",
    y="Revenue",
    palette="deep"
)

plt.title("Average Revenue by Platform Tier")
plt.xlabel("Platform Tier")
plt.ylabel("Average Revenue")

plt.show()
```

```python
import pandas as pd
import plotly.express as px

data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Traffic": [1200, 1450, 1100, 1300, 1700, 2200, 1900],
    "Revenue": [12000, 15500, 10500, 14000, 19500, 26000, 21000],
    "Platform_Tier": ["Basic", "Basic", "Standard", "Standard", "Premium", "Premium", "Premium"]
}

df = pd.DataFrame(data)

fig = px.scatter(
    df,
    x="Traffic",
    y="Revenue",
    color="Platform_Tier",
    title="Traffic vs Revenue"
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Traffic": [1200, 1450, 1100, 1300, 1700, 2200, 1900],
    "Revenue": [12000, 15500, 10500, 14000, 19500, 26000, 21000],
    "Platform_Tier": ["Basic", "Basic", "Standard", "Standard", "Premium", "Premium", "Premium"]
}

df = pd.DataFrame(data)

fig = go.Figure()

fig.add_trace(
    go.Bar(
        x=df["Day"],
        y=df["Traffic"],
        name="Traffic"
    )
)

fig.update_layout(
    title="Traffic Trends Over 7 Days",
    xaxis_title="Day",
    yaxis_title="Traffic",
    template="plotly_white"
)

fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=True)

fig.show()
```
